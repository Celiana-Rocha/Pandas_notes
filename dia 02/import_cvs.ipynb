# diquinha do Téo
se formos fazer um filtro para depois alterar os dados FAÇA UMA CÓPIA (caso nao peceba o pq do erro se tiver SLINCI foi isso)
no codigo abaixo se ele nao utilizasse o copy() ele estaria tentanto modificar a tabela original, pois no python ele nao retorna uma copia logo de cara  ele apenas aponta para os valores ja existentes, e nao é possivel modicar uma tabela original de cara assim utilizando filtros, porem da certo fazendo uma copia dos dados solicitados.

```python
condicao = (df_custumers['Points'] >= 1000) & (df_custumers['Points'] <= 2000)
df_1000_2000 = df_custumers[condicao].copy()

df_1000_2000['Points'] = df_1000_2000['Points'] + 1000
df_1000_2000
```

# Como organizar colunas no DataFrame
usando .columns e .tolist()
que funciona da seguinte forma, passamos o dataframe e pegamos todas as colunas com .columns e após isso transfromamos em uma lista usando .tolist()
 E para utilizar de froma alfabetica use o .sort() que reordena os nomes 

```python
colunas = df_custumers.columns.tolist()
colunas.sort()
df_custumers = df_custumers[colunas]
df_custumers
```

# como renomaer o nome da coluna?
podemos passar o dataframe e ultilizar o metodo rename e como argumento dele outro metodo que tras as colunas o .columns e como argumento dele um dicionario com a chave sendo o que queremos renomear e o valor atribuido a mudança. um ponto importante sobre o rename é que ele retorna um novo dataframe, entao a froma de de fato renomear o DF original é atribuindo a ele mesmo.

```python
df_custumers = df_custumers.rename(columns = {'Name': 'Nome',
                               'Points': 'Pontos'})
df_custumers
```
