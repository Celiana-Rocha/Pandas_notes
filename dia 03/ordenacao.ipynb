```python
import pandas as pd
```

```python
df = pd.read_csv('../datas/customers.csv', sep = ';')
df
```

como ordenar 
-
 - use sort_values(by= 'coluna') assim dessa forma é possivel ordernar pela coluna escolhida por padrao vem do menor para o maior
 - para vir do maior para o menor use ascending = False

```python
df.sort_values(by = 'Points', ascending = False, inplace = True)
df
```

uma forma de fazer um "pipeline" / arrumar os dados
-
 - Voce pode passar tanto o sort_values e ascending = para ordenar os dados pela ordem desejada, como tambem o .rename para renomear se necesario os nomes das colunas

```python
df.sort_values(by = 'Points', ascending = False, inplace = True)
df.rename(columns = {'Name':'Nome', 'Points':'Pontos'}, inplace = True)
df
```

Forma comum de se ver um pipeline
-
 + obs: em pipeline nao pode ter inplace pois ele nao retorna uma DF novo ele altera o original
    + ### uma forma comum de se ver por ai é com a logica de que o primeiro df retorna um DF com isso ja podemos por nele mesmo o nosso rename pois o resutado dele que sera mutado, outro ponto é que ao inves de por inplace podemos atribuir ao df de uma vez. 

```python
df = df.sort_values(by = ['Points', 'Name'], ascending = [False, True]).rename(columns = {'Name':'Nome', 'Points':'Pontos'})
df
```

coisa boa sintax de SORT_VALUES
-

quando passamos o by = nós podemos dar o valor dele como uma lista e no ascending a mesma coisa, a parte boa é que referente a lista o primeiro valor de by vai corresponder o primeiro valor de ascending e assim por diante.

- no caso aqui estamos passando a colu a points e trazendo ela de forma do maior para o menor e a name em ordem alfabetica (na logica normal do menor para o maior)


```python
df.sort_values(by = ['Points', 'Name'], ascending = [False, True]).tail(10)
```
