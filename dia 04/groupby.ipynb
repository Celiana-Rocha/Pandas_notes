explicacao do codigo abaixo

- passamos o df no groupby ( que agrupa por uma determinada coluna especificada) agrupamos pela coluna IdCustomer e pedimos para que pela ordem de IdCustomer faca o .sum(soma) da coluna points de cada usuario.
- podemos transformar a serie obtida em um dataFrame usando .reset_index()

```python
df_sumary = df.groupby(['IdCustomer'])['Points'].sum()
df_sumary.reset_index()
```

mas e se quisermos fazer maic coisas do só somar?
-

explicando o codigo abaixo

- agrupamos por IdCustomer, usamos .agg como metodo para usarmos mais metodos, nele passamos um dicionario com as informacoes, cada titulo corresponde ha uma coluna e seu conteudo o tipo de metodo.

```python
df.groupby(['IdCustomer']).agg({'Points': 'sum', 
                                'UUID': 'count',
                                'DtTransaction': 'max',
                                }).reset_index().rename(columns={'UUID': 'Frequencia',
                                                         'Points': 'valor',
                                                         'DtTransactions': 'UltimaData'})
```
